### Lexical richness
vocabulary = different words in a texts

lexical richness = len(vocabulary) / len(text)

```python
vocabulary = sorted(set(text3))
lex_richness = len(vocabulary) / len(text3)
lex_richness
print(f'Lexical richness: {lex_richness},\n' \
      f'i.e. the number of distinct words is ~{round(lex_richness * 100)}% ' \
       'of the total number of words in the text')
```

Calculate the average number of times each word is used

```python
avg_times_word_used = len(text3) / (len(text3) * lex_richness)
print(f'Each word is used an average of {round(avg_times_word_used, 2)} times')
```

### Frequency Distribution
How can we identify the words of a text that are most informative about a topic or genre of the text?

```python
# text1 - Moby Dick
freq_dist = FreqDist(text1)
NUM_MOST_COMMON = 50
WORD = 'whale'
print(f'The {NUM_MOST_COMMON} tokens in the text are {freq_dist.most_common(NUM_MOST_COMMON)}')
print(f'The word "{WORD}" appears {freq_dist[WORD]} time(s) in the text')
print(freq_dist)
freq_dist.plot(NUM_MOST_COMMON, cumulative=True)
```

The plot above is not very helpful in interpreting the text as it these small words do not offer any context of the text.

Use hapaxes words (words that occur only once in the text) to try provide more context.

HINT: this doesn't help either

```python
freq_dist.hapaxes()
```

### Fine-grained selection of words
Use fine-grained selection of words to provide context of a text


We only want to select words (w) in the vocabulary (V) where that word is longer than a specified length (L)

${ w\;|\;w\,∈\,V\;and\;len(w)\,>\,L}$

We'll see that this is not very helpful either as most of these long words are also hapaxes words.

```python
V = sorted(set(text1))
L = 15

words = sorted([w for w in V if len(w) > L])
words
```

We can improve on this, let's search for words (w) of a defined length (L) in the text that have a defined frequency (F).

${ w\,|\,w\;∈\,V\;and\;len(w)\,>\,N\;and\;f(w)\;>\;F }$

```python
freq_dist = FreqDist(text1)
L = 7
F = 7

words = sorted([w for w, f in freq_dist.items() if len(w) > L and f > F])
words
```

### Collocations
A specific pattern of words that occur together with a higher frequency than would be expected by chance.

e.g. "red wine"

Another characteristic is their resistance to substitution with similar words. For example, "maroon wine" sounds strange and so, the collocation "red wine" is resistant to substitution.
