```python
import nltk
from urllib import request
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.bbc.com/news/world-australia-66220494'
html_article = request.urlopen(URL).read().decode('utf8')
raw_article = BeautifulSoup(html_article, 'html.parser').get_text()
```

```python
tokens = nltk.word_tokenize(raw_article)
```
