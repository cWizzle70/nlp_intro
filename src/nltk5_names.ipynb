```python
import nltk
from nltk.corpus import names, cmudict, swadesh
```

#### Names Corpus

```python
female_names = names.words('female.txt')
male_names = names.words('male.txt')
```

```python
ambiguous_names = [w for w in names.words() if w in female_names and w in male_names]
ambiguous_names
```

#### Prove the majority of names ending in 'a' are female

```python
cfd = nltk.ConditionalFreqDist(
    (file_id, name[-1])
    for file_id in names.fileids()
    for name in names.words(file_id)
)
cfd.plot()
```
